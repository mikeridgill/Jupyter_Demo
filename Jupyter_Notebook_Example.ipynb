```python
def least_difference(a, b, c):
    """Return the smallest difference between any two numbers
    among a, b and c.
    
    >>> least_difference(1, 5, -5)
    4
    """
    diff1 = abs(a - b)
    diff2 = abs(b - c)
    diff3 = abs(a - c)
    return min(diff1, diff2, diff3)
```

```python
least_difference(1, 10, 100)
```

#### Complicated example

Using the station gauge locations from the Pandas example above, let's plot their locations.  This will require additional packages and data, plus use of functions, for loops and conditionals.

```python
# Reminder of the data
station_locations.head(10)
```

##### Calculate latitude and longitude for each station
This will require using a package called [OSGridConverter](https://pypi.org/project/OSGridConverter/).

```python
from OSGridConverter import grid2latlong
station_locations['latitude'] = 0
station_locations['longitude'] = 0

for i in range(len(station_locations)):
    grid_ref = station_locations.loc[station_locations.index[i], 'GRID_REF']
    lat, lon = grid2latlong(grid_ref,  tag='WGS84').latitude, grid2latlong(grid_ref,  tag='WGS84').longitude
    station_locations.loc[station_locations.index[i], 'latitude'] = lat
    station_locations.loc[station_locations.index[i], 'longitude'] = lon
    
station_locations.head()
```

...but wait, the shapefile we have for Wales uses BNG coordinates.  Is there a package to convert these lats and lons into BNG?  Of course: [bng-latlon](https://pypi.org/project/bng-latlon/).

```python
from latlon_to_bng import WGS84toOSGB36
station_locations['bng_x'] = 0
station_locations['bng_y'] = 0

for i in range(len(station_locations)):    
    station_locations.loc[station_locations.index[i], 'bng_x'] = WGS84toOSGB36(station_locations.loc[station_locations.index[i], 'latitude'], station_locations.loc[station_locations.index[i], 'longitude'])[0]
    station_locations.loc[station_locations.index[i], 'bng_y'] = WGS84toOSGB36(station_locations.loc[station_locations.index[i], 'latitude'], station_locations.loc[station_locations.index[i], 'longitude'])[1]

station_locations.head()
```

##### Import shapefiles for Wales boundary and rivers
This will require using the package [shapefile](https://pypi.org/project/pyshp/).

```python
import shapefile
rivers_sf = shapefile.Reader("./data/mainRiversLle/NRW_MAIN_RIVERSLine")
wales_sf = shapefile.Reader("./data/walesBoundary/Wales_boundary")
```

##### Plot of Wales showing main rivers and the locations of gauges

```python
plt.figure(figsize=(14,14))
for shape in rivers_sf.shapeRecords():
    x = [i[0] for i in shape.shape.points[:]]
    y = [i[1] for i in shape.shape.points[:]]
    plt.plot(x,y, color="gray", linewidth=0.4)
for shape in wales_sf.shapeRecords():
    for i in range(len(shape.shape.parts)):
        i_start = shape.shape.parts[i]
        if i==len(shape.shape.parts)-1:
            i_end = len(shape.shape.points)
        else:
            i_end = shape.shape.parts[i+1]
        x = [i[0] for i in shape.shape.points[i_start:i_end]]
        y = [i[1] for i in shape.shape.points[i_start:i_end]]
        plt.plot(x,y, color="black", linewidth=0.8)

station_plots = plt.scatter(station_locations['bng_x'], station_locations['bng_y'], label='Gauging station')
    
for index, row in station_locations.iterrows():
    plt.annotate(row['STATION_NUMBER'], (row['bng_x'], row['bng_y']))

plt.xlabel('Eastings (BNG)')
plt.ylabel('Northings (BNG)')
plt.legend(handles=[station_plots])
plt.grid()
plt.show()
```
